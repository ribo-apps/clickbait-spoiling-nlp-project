```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Example data
names = ['tf-idf','roberta', 'falcon', 'gpt-3.5', 'alpaca', 't5 large']
values = [0.0183509, 2.94e-157, 0.01189, 0.14789, 0.005389, 0.19242]
values = np.log(np.array(values) * 100000)

# Create bar plot using Seaborn
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x=names, y=values, ax=ax)

# Set plot title and labels
ax.set_title('LM Comparisons')
ax.set_xlabel('Model')
ax.set_ylabel('BLEU Score (log scale)')
ax.set_yscale('log')

# Display plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
names = ['tf-idf', 'roberta', 'falcon', 'gpt-3.5', 'alpaca', 't5 large']
values = [0.7890071868896484, 0.792813, 0.7621, 0.86662, 0.8215, 0.8644]

# Find the index of the best value
best_index = values.index(max(values))

# Create bar plot using Seaborn
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x=names, y=values, ax=ax, palette='Set2')

# Highlight the best value with a red line
ax.axhline(y=values[best_index], color='red', linestyle='--')

# Set plot title and labels
ax.set_title('LM Comparisons')
ax.set_xlabel('Model')
ax.set_ylabel('BERT Score')

# Display plot
plt.show()
```
